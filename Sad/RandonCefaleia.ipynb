```python
def GerarRandonList(posIni,posFin):
    listaRandon=[]
    rangeNumeros = 5000
    
    for i in range(0,rangeNumeros):
        n = random.randint(posIni,posFin)
        listaRandon.append(n)

    return listaRandon
```

```python
def GerarRandonGenero(listPadrao):
    listaRandon=[]
    rangeNumeros = 5000

    for i in range(0,rangeNumeros): 
        n = random.choice(listPadrao)
        listaRandon.append(n)
    
    return listaRandon 
```
